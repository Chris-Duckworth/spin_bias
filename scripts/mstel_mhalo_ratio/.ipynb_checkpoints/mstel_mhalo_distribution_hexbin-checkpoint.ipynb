### mstel_mhalo_distribution
- This script looks at the distribution of mstel vs mhalo and then colours on various properties.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid

import catalog
import catalog_init 
import catalog_process
import catalog_plot
import copy

import imp 
imp.reload(catalog)
imp.reload(catalog_plot)
imp.reload(catalog_process)
imp.reload(catalog_init)

# plotting defaults.
from matplotlib.pyplot import rcParams
catalog_plot.default(rcParams)

plt.style.use("dark_background")
```

```python
basepath = '/Users/chrisduckworth/astronomy/projects/spin_bias/catalogues/'
outpath = '/Users/chrisduckworth/astronomy/projects/spin_bias/plots/mstel_mhalo_distributions/mpl9/'

# initiating catalog class objects.
total = catalog.Catalog(basepath, version='mpl9', match_to_lim=True)
# matching to cw definitions and selecting centrals only.
total.match_to_cw(basepath, version='mpl9', sigma=3)
total.remove_satellites(group_cat='lim') 

# creating copies of these objects and cutting on morphology.
ltgs = copy.deepcopy(total) 
ltgs.select_morphology('ltg')

etgs = copy.deepcopy(total)
etgs.select_morphology('etg')

unclassified = copy.deepcopy(total)
unclassified.select_morphology('unclassified')
```

```python
# finding rolling mean for populations.
fraction = 0.1 # rolling mean contains 5% of total points.

p_total = total.compute_expected_y('log_halo_mass', 'nsa_elpetro_mass_logscale', method="running_mean", return_plot=True, plot_x_extent=[11, 15], plot_y_extent=[8.5, 11.5], n_neighbours=int(fraction*total.df.shape[0]))
p_etgs = etgs.compute_expected_y('log_halo_mass', 'nsa_elpetro_mass_logscale', method="running_mean", return_plot=True, plot_x_extent=[11, 15], plot_y_extent=[8.5, 11.5], n_neighbours=int(fraction*etgs.df.shape[0]))
p_ltgs = ltgs.compute_expected_y('log_halo_mass', 'nsa_elpetro_mass_logscale', method="running_mean", return_plot=True, plot_x_extent=[11, 15], plot_y_extent=[8.5, 11.5], n_neighbours=int(fraction*ltgs.df.shape[0]))
p_unclassified = unclassified.compute_expected_y('log_halo_mass', 'nsa_elpetro_mass_logscale', method="running_mean", return_plot=True, plot_x_extent=[11, 15], plot_y_extent=[8.5, 11.5], n_neighbours=int(fraction*unclassified.df.shape[0]))

x_total = np.linspace(total.df.log_halo_mass.min(), total.df.log_halo_mass.max())
x_etgs = np.linspace(etgs.df.log_halo_mass.min(), etgs.df.log_halo_mass.max())
x_ltgs = np.linspace(ltgs.df.log_halo_mass.min(), ltgs.df.log_halo_mass.max())
x_unclassified = np.linspace(unclassified.df.log_halo_mass.min(), unclassified.df.log_halo_mass.max())
```

### Making a series of row plots that show the distributions coloured on a property split into morphs.

1) Stellar to halo ratio

```python
# Set up figure and image grid
fig= plt.figure(figsize=(40, 5))

ax = ImageGrid(fig, 111,          # as in plt.subplot(111)
                 nrows_ncols=(1,4),
                 axes_pad=0.1,
                 share_all=True,
                 cbar_location="right",
                 cbar_mode="single",
                 cbar_size="7%",
                 cbar_pad=0.15,
                 )

extent = [11, 15, 8.5, 11.8]
vrange = [-3.5, -1.5]
grid = 40 
mincnt = 0

hb = total.plot_hexbin('log_halo_mass', 'nsa_elpetro_mass_logscale', ax[0], C=total.df.stellar_to_halo_ratio.values, mincnt=mincnt, gridsize=grid, extent=extent, cmap='viridis', vmin=vrange[0], vmax=vrange[1])
hb = etgs.plot_hexbin('log_halo_mass', 'nsa_elpetro_mass_logscale', ax[1], C=etgs.df.stellar_to_halo_ratio.values, mincnt=mincnt, gridsize=grid, extent=extent, cmap='viridis', vmin=vrange[0], vmax=vrange[1])
ltgs.plot_hexbin('log_halo_mass', 'nsa_elpetro_mass_logscale', ax[2], C=ltgs.df.stellar_to_halo_ratio.values, mincnt=mincnt, gridsize=grid, extent=extent, cmap='viridis', vmin=vrange[0], vmax=vrange[1])
unclassified.plot_hexbin('log_halo_mass', 'nsa_elpetro_mass_logscale', ax[3], C=unclassified.df.stellar_to_halo_ratio.values, mincnt=mincnt, gridsize=grid, extent=extent, cmap='viridis', vmin=vrange[0], vmax=vrange[1])

# adding expected mean line.
ax[0].plot(x_total, p_total(x_total), color='w') 
ax[1].plot(x_etgs, p_etgs(x_etgs), color='w') 
ax[2].plot(x_ltgs, p_ltgs(x_ltgs), color='w') 
ax[3].plot(x_unclassified, p_unclassified(x_unclassified), color='w') 

# Colorbar
cb = ax[0].cax.colorbar(hb)
cb.set_label_text('$\mathrm{M_{stel} / M_{halo}}$', fontsize=14)
ax[0].cax.toggle_label(True)

titles = ['All galaxies', 'ETGs', 'LTGs', 'Unclassified']

for ind, axis in enumerate(ax):
    axis.set_xlabel('$\mathrm{log_{10}(M_{halo}) \; [M_{\odot}]}$', fontsize=14)
    axis.set_title(titles[ind], fontsize=14)
    
ax[0].set_ylabel('$\mathrm{log_{10}(M_{stel}) \; [M_{\odot}]}$', fontsize=14)
ax[0].set_ylim([8.5, 11.5])
ax[0].set_xlim([11, 14.5])

plt.savefig(outpath+'mstel_mhalo_c-ratio-double-bin.pdf', format='pdf', bbox_inches='tight')
```
