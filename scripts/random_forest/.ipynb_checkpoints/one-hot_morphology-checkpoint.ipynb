# one-hot_morphology
- finding the best way to include morphology as a feature for input into the random forest
- testing out one-hot encoding for individual catagories
- alternatively using vote fraction for disk / features

```python
import pandas as pd 
import catalog
import catalog_init
import catalog_process
import numpy as np 
import matplotlib.pyplot as plt
```

```python
basepath = '../../catalogues/'
outpath = './'

# initiating catalog class objects.
total = catalog.Catalog(basepath, version='mpl9', match_to_lim=True)
# matching to cw definitions and selecting centrals only.
total.match_to_cw(basepath, version='mpl9', sigma=3)
```

```python
list(total.df.columns)
```

### one-hot encoding various morphological classifications

```python
morph = np.full(total.df.shape[0], 'unclassified')
morph[catalog_init.select_morphology(total.df, morphology='etg', vote_frac=0.7).index] = 'etg'
morph[catalog_init.select_morphology(total.df, morphology='s0sa', vote_frac=0.7).index] = 's0sa'
morph[catalog_init.select_morphology(total.df, morphology='sbsd', vote_frac=0.7).index] = 'sbsd'

plt.hist(morph);
```

```python
# adding columns to feature pandas and encoding.

features = total.df[['nsa_elpetro_mass_logscale', 
                        't01_smooth_or_features_a01_smooth_debiased',
                        'nsa_elpetro_ba',
                        'nsa_sersic_n',
                        'log_halo_mass', 
                        'log_dnode_norm', 
                        'log_dskel_norm',
                        'log_dwall_norm']]

features['morph'] = morph
features
```

```python
features = pd.get_dummies(features)
features.head(5)
```
